```python
# process df_merged
df_merged = pd.read_csv(
    "/Users/lullu/study/meal_manager_server/data/DB_merged.csv", encoding="utf-8-sig"
)

# 제조사 이름에 (주), ㈜ 제거
df_merged["제조사/유통사"] = df_merged["제조사/유통사"].str.replace("(주)", "")
df_merged["제조사/유통사"] = df_merged["제조사/유통사"].str.replace("㈜", "")
df_merged["제조사/유통사"] = df_merged["제조사/유통사"].str.strip()
print(df_merged.shape[0])

# 모든 column의 value 중 1g미만 -> 0
for column in df_merged.columns:
    # if value contains "미만", replace with 0
    if df_merged[column].dtype == "object":
        df_merged.loc[df_merged[column].str.contains("미만", na=False), column] = 0


# 틴단지 모두 있는것만 남기기
df_merged = df_merged[df_merged["탄수화물(g)"] != "-"]
df_merged = df_merged[df_merged["지방(g)"] != "-"]
df_merged = df_merged[df_merged["단백질(g)"] != "-"]
print(df_merged.shape[0])

# approximate calories with carbs, fat, protein, if calories are missing
df_merged["탄수화물(g)"] = df_merged["탄수화물(g)"].astype(float)
df_merged["지방(g)"] = df_merged["지방(g)"].astype(float)
df_merged["단백질(g)"] = df_merged["단백질(g)"].astype(float)
df_merged.loc[df_merged["에너지(㎉)"] == "-", "에너지(㎉)"] = (
    df_merged["탄수화물(g)"] * 4 + df_merged["지방(g)"] * 9 + df_merged["단백질(g)"] * 4
)

# only g, mL
df_merged = df_merged[df_merged["내용량_단위"].isin(["g", "mL"])]
# print unique values of "내용량_단위", and count of each value
print(df_merged["내용량_단위"].value_counts())


df_merged.to_csv(
    "/Users/lullu/study/meal_manager_server/data/DB_merged_processed.csv", index=False
)
```
